```python
import pandas as pd
import matplotlib.pyplot as  plt
plt.style.use('seaborn-dark')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/muga01/python_basics/main/train.csv')
```

```python
data.head()
```

```python
data.drop(['target_benzene', 'target_nitrogen_oxides'], axis=1, inplace=True )
```

```python
data.head()
```

```python
dy = pd.DatetimeIndex(data['date_time'])
period_d = dy.to_period('D')
period_d
```

MONOCABORN DATA

```python
data_per_day = data.groupby(period_d).sum()
carbon = data_per_day['target_carbon_monoxide']
carbon1 = data_per_day['target_carbon_monoxide'].sort_values(ascending=False)
carbon1
```

```python
#pd.to_datetime('2010-12-24').to_period('H').hour
pd.to_datetime('2010-12-24').strftime('%W')
pd.to_datetime('2010-12-24').strftime('%w')
pd.to_datetime('2010-12-24').strftime('%c')
pd.to_datetime('2011-01-01').strftime('%c')
```

```python
plt.style.use('seaborn-dark',)
carbon.plot(kind='area',figsize=(8,4),legend=True)
plt.grid()
plt.legend(loc='upper center')
```

TEMPERATURE

```python
temp = data_per_day['deg_C']
temp1 = data_per_day['deg_C'].sort_values(ascending=False)
temp1
```

```python
plt.style.use('seaborn-dark',)
temp.plot(kind='area',figsize=(8,4),legend=True,cmap='RdGy',title='TEMPERATURE',ylabel='Deg_C')
plt.grid()
plt.legend(loc='upper right')
```

```python
pd.to_datetime('2010-07-22').strftime('%W')
pd.to_datetime('2010-07-22').strftime('%c') # Max Thu Jul 22 00:00:00 2010
pd.to_datetime('2011-01-01').strftime('%c')#  min  Sat Jan  1 00:00:00 2011
```

TATHMINI 2

```python
date_index = pd.DatetimeIndex(data['date_time'])
dta = date_index.to_period('H')
dta
```

```python
days = ['monday','tuesday','wednesday','thursday','friday','saturday','sunday']
plt.style.use('dark_background')
p = data.pivot_table(index=dta.dayofweek,aggfunc=({'target_carbon_monoxide':'mean',
                                                   'absolute_humidity':'mean'}))
p.index=days
p.plot()
plt.xticks(rotation=90);                                   
```

```python
plt.style.use('dark_background')
p = data.pivot_table(index=dta.month,aggfunc=({'target_carbon_monoxide':'mean',
                                                   'absolute_humidity':'mean'}))

p.plot()
plt.xticks(rotation=360); 
```

```python
plt.style.use('dark_background')
p = data.pivot_table(index=dta.year,aggfunc=({'target_carbon_monoxide':'mean',
                                                   'sensor_1':'mean'})).plot()
plt.xticks(rotation=360); 
```

```python
plt.style.use('dark_background')
p = data.pivot_table(index=dta.month,aggfunc=({'deg_C':'mean',
                                                   'sensor_5':'mean'})).plot()
plt.xticks(rotation=360); 
```
