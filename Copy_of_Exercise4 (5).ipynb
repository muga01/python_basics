```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/muga01/python_basics/main/train.csv')
data.head()
```

```python
data.drop(['date_time','target_benzene','target_nitrogen_oxides'], axis=1, inplace=True)
```

```python
from itertools import combinations
import matplotlib.pyplot as plt
```

```python
data.head(1)
```

```python
def chart_plot(x,y): 
    plt.style.use('seaborn-white')
    fig = plt.figure(figsize=(1,2))
    grid = plt.GridSpec(5,4)
    a = fig.add_subplot(grid[0:4,:1])
    b = fig.add_subplot(grid[:4,1:])
    c = fig.add_subplot(grid[4:,1:])
    a.hist(data[str(y)],orientation='horizontal',color='Gray' ,histtype='stepfilled',edgecolor='red',bins = 40)
    b.scatter(data[str(y)],data[str(x)], color='black',alpha=0.3)
    c.hist(data[str(x)],orientation='vertical',color='Gray',histtype='stepfilled',edgecolor='red',bins = 40);
    # plt.legend(loc='upper right');
    c.set(xlabel= data[str(x)].name)
    a.set(ylabel= data[str(y)].name)
```

```python
chart_plot('deg_C','sensor_3')
```

```python
import matplotlib.pyplot as plt
from itertools import combinations
def graph_charts(x,y):
   plt.style.use('seaborn-white')
   t = list(combinations(data,2));
   for i in range(0,36):
       x= t[i][0];
       y = t[i][1];
       fig = plt.figure(figsize=(4,5))
       grid = plt.GridSpec(5,4,hspace=0.4,wspace=0.2, )
       a = fig.add_subplot(grid[0:4,:1])
       b = fig.add_subplot(grid[:4,1:])
       c = fig.add_subplot(grid[4:,1:])
       a.hist(data[str(y)],orientation='horizontal',color='Gray' ,histtype='stepfilled',edgecolor='red',bins = 40);
       b.scatter(data[str(y)],data[str(x)], color='black',alpha=0.3)
       c.hist(data[str(x)],orientation='vertical',color='Gray',histtype='stepfilled',edgecolor='red',bins = 40);
       # plt.legend(loc='upper right');
       c.set(xlabel= data[str(x)].name)
       a.set(ylabel= data[str(y)].name)
       plt.figure()

```

```python
graph_charts('deg_C','sensor_1');
```
